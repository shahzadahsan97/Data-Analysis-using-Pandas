```python
import pandas as pd
import numpy as np
from pandas import DataFrame
np.random.seed(5)

students = pd.DataFrame({'Name':['Ahsan' , 'Ali' , 'Umer' , 'Bilal' , 'Ahmed'],
                        'age':np.random.randint(5,25 , size  = [5]),
                        'Percent' : np.random.randint(40,90,size = [5])})
```

```python
students
```

```python
def name(age):
    if age < 18:
        return 'Non-Adult'
    
    else:
        return 'Adult'
```

```python
students['Status'] = students['age'].map(name)
```

```python
students
```

```python
def checkPercentage(Percent):
    if Percent >= 50:
        return 'Pass'
    else:
        return 'Fail'
```

```python
students['newResult'] = students['Percent'].map(checkPercentage)
students
```
