```python
df.Marks.fillna(method = 'ffill' , inplace = True)
df
```

```python
df = pd.DataFrame({'Names':['Hamza','Ali','Shahzad',
                            'Ali', 'Gulraiz']
                   ,'Marks':[30,30,80,np.nan,50]
                    ,'Age':[40,50,34,np.nan,np.nan]})
df
```

```python
df.duplicated(['Age'])
```

```python
display(df)
df.drop_duplicates(['Names'])
```

```python
display(df)
df.drop_duplicates(['Names'], keep = 'last')
```

```python
name_to_lname = {'Shahzad':'Ahsan',
                 'Hamza':'Alee', 
                 'Ali':'Khan'}
```

```python
df['Fathers Name'] = df.Names.map(name_to_lname)
df
```

```python
def age(r):
    if(r['Age'] < 18):
        return "Child"
    else:
        return "Adult"
    
df["Status"] = df.apply(age, axis = 1)
```

```python
df
```

```python
df = pd.DataFrame({'Names':['Hamza','Ali','Shahzad',
                            'Ali', 'Hamza']
                   ,'Age':[15,10,80,14,50]
                    ,'Marks':[40,54,65,35,89]})
df
```

```python
df['Grade']=df['Marks'].apply \
            (lambda x:"Pass" if x>60 else "Fail")
```

```python
df
```

```python
df['Status'] = df['Age'].apply \
                (lambda a: 'Adult' if a>=18 else 'Non Adult') 
```

```python
df
```
