```python
df = pd.read_csv('data for panda exercise.csv')
df
```

```python
def func(x, y):
    if pd.isnull(x):
        return y
    else:
        return x
df['Quiz 1'] = df['Quiz 1'].apply(func,args=[1])
```

```python
df["Status"] = df.Grade.apply( \
                    lambda x: "Fail" if x == 'F' else "Pass" )
```
